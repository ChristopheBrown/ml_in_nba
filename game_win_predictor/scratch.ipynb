```python
from svm_game_win_predictor import *
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
```

```python
# defaults: C=1.0 gamma='scale', degree=3
kernels = ['rbf', 'poly']
gammas = [0.1, 1, 10, 100]
cs = [0.1, 1, 10, 100, 1000]
degrees = [0, 1, 2, 3, 4, 5, 6]

league_accuracies = {}
league_wins_avg = {}
```

```python
%%time

# Vanilla LinearSVC
league_accuracies['vanilla_linearsvc'] = run(nba, printouts=False)
```

```python
%%time

# Vanilla SVC
league_accuracies['vanilla_svc'] = run(nba, model=SVC(random_state=0), printouts=False)
```

```python
%%time

# Tuned SVC - rbf kernel
model = SVC(kernel=kernels[0], random_state=0)
league_accuracies['vanilla_svc_rbf'] = run(nba, model=model, printouts=False)
```

```python
%%time

# Tuned SVC - poly kernel
model = SVC(kernel=kernels[1], random_state=0)
league_accuracies['vanilla_svc_poly'] = run(nba, model=model)
```

```python
%%time

# Tuned SVC - rbf kernel + gammas
for gamma in gammas:
    print(f'gamma = {gamma}')
    model = SVC(kernel=kernels[0], random_state=0, gamma=gamma)
    league_accuracies[f'svc_rbf_gamma_{gamma}'] = run(nba, model=model, printouts=False)
```

```python
%%time

# Tuned SVC - poly kernel
for gamma in gammas:
    print(f'gamma = {gamma}')
    model = SVC(kernel=kernels[1], random_state=0, gamma=gamma)
    league_accuracies[f'svc_poly_gamma_{gamma}'] = run(nba, model=model, printouts=False)
```

```python
%%time

# Tuned SVC - rbf kernel + C
for c in cs:
    print(f'c = {c}')
    model = SVC(kernel=kernels[0], random_state=0, C=c)
    league_accuracies[f'svc_rbf_c_{c}'] = run(nba, model=model, printouts=False)
```

```python
%%time

# Tuned SVC - poly kernel + C
for c in cs:
    print(f'c = {c}')
    model = SVC(kernel=kernels[1], random_state=0, C=c)
    league_accuracies[f'svc_poly_c_{c}'] = run(nba, model=model, printouts=False)
```

```python
%%time

# Tuned LinearSVC - C
for c in cs:
    print(f'c = {c}')
    model = LinearSVC(random_state=0, C=c)
    league_accuracies['linearsvc_c'] = run(nba, model=model, printouts=False)
```
